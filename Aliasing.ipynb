```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts=0.001
omegas = 2*np.pi/ts
omega0 = 0.6*omegas
```

```python
fig = plt.figure(figsize=(10,3))
t = np.linspace(0,0.010,1000)
td = np.arange(0,.01,.001)
plt.stem(td,np.cos(omega0*td))
plt.plot(t,np.cos(omega0*t),'r--')
plt.plot(t,np.cos(0.4*omegas*t),'g--')
plt.grid()
```

```python
fig = plt.figure(figsize=(10,3))
t = np.linspace(0,0.010,1000)
td = np.arange(0,.01,.001)
plt.plot(t,np.cos(omega0*t))
plt.stem(td,np.ones(len(td)))
plt.grid()
```
