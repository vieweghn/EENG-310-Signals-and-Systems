```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

image.png

```python
t = np.linspace(0, 5, 100)
dt = t[1]-t[0]

def u(t):
  return np.heaviside(t, 0)

def nu(t):
  return -10*np.exp(-10*t)*u(t)

def x(t):
  return u(t)

# 1a.
y = signal.convolve(nu(t), x(t), mode='full', method='fft')*dt

plt.plot(t, x(t), label='x(t)')
plt.plot(t, nu(t), label='nu(t)')
plt.plot(t, y[:len(t)], label='y(t)')
plt.xlim(0, 2)
plt.grid()
plt.legend()
plt.xlabel('t')
```

```python
# 1b.
v = 2*np.exp(-2*t)*u(t)
x = u(t)

y = signal.convolve(2*np.exp(-2*t)*u(t),u(t), mode='full', method='fft')*dt
plt.plot(t,y[:len(t)], label='y(t)')
plt.plot(t, v, label='v(t)')
plt.plot(t, x, label='x(t)')
plt.legend()
plt.grid()
```

# 1c.
The response from part a is higher than in part b. The response in part a reaches its maximum voltage faster than in part b.

```python
# 1d.
v = 2*np.exp(-2*t)*u(t)
x = u(t-1)

y = signal.convolve(v,x, mode='full', method='fft')*dt
ypa = signal.convolve(2*np.exp(-2*t)*u(t),u(t), mode='full', method='fft')*dt
yp = ypa[len(t)-1:]
plt.plot(t,y[:len(t)], label='y(t)')
plt.plot(t, v, label='v(t)')
plt.plot(t, x, label='x(t)')
plt.plot(t, yp[:len(t)], label="y'(t)")
plt.xlim(0, 5)
plt.legend()
plt.grid()
```
