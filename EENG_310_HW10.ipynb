A. (v44) The current through a $10\mathrm{~k\Omega}$ resistor is given by a triangular waveform with $A=4\mathrm{~mA}$ and $T=0.2\mathrm{~s}$.

1. Determine the exact average power consumed by the resistor.
2. Using only the first four non-zero terms of the Fourier Series, estimate the average power.
3. What is the percent difference between (1) and (2)?

(Ans: 53.33 mW, first four non-zero terms is 0.03%  different from the exact result.)

image.png

HW1010Aâ‹…1.png

```python
R = 10e3
A = 4e-3
T0 = .2

avgPower = (R * A**2/3)*1000 # in mW
print(f'1.) Pavg = {avgPower:.2f} mW')

# xt = 8*A / n**2 / np.pi**2 * np.cos(2*n*np.pi*t / T0)
n = 1
Px = 0
a0 = 0
an = 0
bn = 8*A / n**2 / np.pi**2
for n in range(1, 8, 2):
  Px = Px + ((an)**2 + (8*A / n**2 / np.pi**2)**2)/2
Px = (Px*R + a0**2) *1000 # in mW
print(f'2.) Px = {Px:.3f} mW')

perDiff = abs((avgPower - Px) / avgPower) * 100
print(f'3.) {perDiff:.2f}% difference')
```
