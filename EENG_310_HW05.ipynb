A. For the circuit below determine a) $H(s)=Vo/Vi$, and b) $h(t)$, given $R_1 = 1\mathrm{~\Omega}$, $R_2=2\mathrm{~\Omega}$, $L_1= 1 \mathrm{~mH}$, $L_2=2\mathrm{~mH}$.

image.png

```python
Vi, Vo, V1, s = sp.symbols('Vi Vo V1 s', real=True, positve=True)
R1 = 1
R2 = 2
ZL1 = 1e-3*s
ZL2 = 2e-3*s

f1 = (V1 - Vi) / R1 + V1/ZL1 + (V1 - Vo) / ZL2
f2 = (Vo - V1) / ZL1 + Vo/R2

solutions = sp.solve([f1, f2], [V1, Vo])
#print(solutions)
Hs = sp.simplify(solutions[Vo] / Vi)
Hs
```

B. For the circuit below determine a) $H(s)=Vo/Vi$ and b) $h(t)$ given that $R_1=1\mathrm{~\Omega}$, $R_2=2\mathrm{~\Omega}$, $C_1=1\mathrm{~\mu F}$, $C_2=2\mathrm{~\mu F}$.

image.png

```python
Vi, Vo, V1, s = sp.symbols('Vi Vo V1 s', real=True, positive=True)
R1 = 1
R2 = 2
C1 = sp.Rational(1, 1e6)
C2 = sp.Rational(1, 2e6)
ZC1 = 1 / (s*C1)
ZC2 = 1 / (s*C2)

f1 = (V1 - Vi) / R1 + (V1 / ZC1) + (V1 - Vo) / ZC2
f2 = (Vo - V1) / ZC2 + (Vo / R2)
solution = sp.solve([f1, f2], [V1, Vo])
#print(solution)
Hs = sp.simplify(solution[Vo] / Vi)
Hs
```
